```python
#write code for above cell and compute the running time and also plot graph.
import time
import random
import matplotlib.pyplot as plt

def selection_sort(arr):
    n = len(arr)
    for i in range(n):
        min_index = i
        for j in range(i + 1, n):
            if arr[j] < arr[min_index]:
                min_index = j
        arr[i], arr[min_index] = arr[min_index], arr[i]

input_sizes = []
running_times = []

sizes = [100, 200, 300, 400, 500, 600, 700, 800, 900, 1000]

for n in sizes:
    numbers = [random.randint(1, 10000) for _ in range(n)]

    start_time = time.time()
    selection_sort(numbers)
    end_time = time.time()

    input_sizes.append(n)
    running_times.append(end_time - start_time)

plt.plot(input_sizes, running_times, marker='o')
plt.xlabel("Number of Elements (n)")
plt.ylabel("Running Time (seconds)")
plt.title("Selection Sort Time Complexity")
plt.grid(True)
plt.show()
```

```python
def counting_sort(arr):
    max_val = max(arr)
    count = [0] * (max_val + 1)

    for num in arr:
        count[num] += 1

    sorted_arr = []
    for i in range(len(count)):
        sorted_arr.extend([i] * count[i])

    return sorted_arr
arr = [4,3,2,1,4,3,2,4,3,4]
print(counting_sort(arr))
```

```python
def bucket_sort(arr):
    if len(arr) == 0:
        return arr

    max_val = max(arr)
    min_val = min(arr)

    bucket_count = max_val - min_val + 1
    buckets = [[] for _ in range(bucket_count)]

    for num in arr:
        buckets[num - min_val].append(num)

    sorted_arr = []
    for bucket in buckets:
        sorted_arr.extend(bucket)

    return sorted_arr


arr = [8,5,3,1,9,6,0,7,4,2,5]
print(bucket_sort(arr))
```

```python
def counting_sort_for_radix(arr, exp):
    n = len(arr)
    output = [0] * n
    count = [0] * 10  # digits 0-9

    for i in range(n):
        index = (arr[i] // exp) % 10
        count[index] += 1

    for i in range(1, 10):
        count[i] += count[i - 1]

    for i in range(n - 1, -1, -1):
        index = (arr[i] // exp) % 10
        output[count[index] - 1] = arr[i]
        count[index] -= 1

    for i in range(n):
        arr[i] = output[i]


def radix_sort(arr):
    max_val = max(arr)
    exp = 1

    while max_val // exp > 0:
        counting_sort_for_radix(arr, exp)
        exp *= 10

    return arr


arr = [18,5,100,3,1,19,6,0,7,4,2]
print(radix_sort(arr))
```

```python
#Prob_4 Take Suitable example and write function that runs in logarithic time. plot graph as well similar to above example.
import math
def binary_search(arr, low, high, target):
    while low <= high:
        mid = (low + high) // 2
        if arr[mid] == target:
            return mid
        elif arr[mid] < target:
            low = mid + 1
        else:
            high = mid - 1
    return -1
n_values = list(range(1, 101))
time_values = []
for n in n_values:
    arr = list(range(n))
    target = -1
    binary_search(arr, 0, len(arr)-1, target)
    time_values.append(math.log2(n))

plt.figure()
plt.plot(n_values, time_values)
plt.xlabel("Input size n")
plt.ylabel("Time")
plt.title("Logarithmic Time Complexity O(log n)")
plt.show()
```

```python
def fibo_rec(n):
    if n < 2:
        return n
    else:
        return fibo_rec(n - 1) + fibo_rec(n - 2)
#compute the time in ms for above cell.
import time
start_time = time.perf_counter()
n = 35
result = fibo_rec(n)
end_time = time.perf_counter()
elapsed_time = (end_time - start_time)*1000
print(f"\nFunction returned: {result}")
print(f"Elapsed time: {elapsed_time} ms")
```

```python
import time

# -----------------------------
# 1. Wiggle Sort
# -----------------------------
def wiggle_sort(nums):
    nums.sort()
    half = (len(nums)+1)//2
    small = nums[:half][::-1]
    large = nums[half:][::-1]

    nums[::2] = small
    nums[1::2] = large
    return nums


# -----------------------------
# 2. Star Pattern
# -----------------------------
def star_pattern(n):
    for i in range(1, n+1):
        for j in range(i):
            print("*", end=" ")
        print()


# -----------------------------
# 3. Reverse Pairs
# -----------------------------
def reverse_pairs(nums):
    count = 0
    for i in range(len(nums)):
        for j in range(i+1, len(nums)):
            if nums[i] > 2 * nums[j]:
                count += 1
    return count


# -----------------------------
# 4. Alphabet Pattern
# -----------------------------
def alphabet_pattern(n):
    for i in range(1, n+1):
        for j in range(i):
            print(chr(64+i), end="")
        print()


# =============================
# MAIN PROGRAM
# =============================

# Wiggle Sort
nums1 = [1,5,1,1,6,4]

start = time.time()
result = wiggle_sort(nums1)
end = time.time()

print("Wiggle Sorted Array:", result)
print("Execution Time:", end-start, "seconds")


print("\nStar Pattern")
start = time.time()
star_pattern(4)
end = time.time()
print("Execution Time:", end-start, "seconds")


nums2 = [1,3,2,3,1]

start = time.time()
pairs = reverse_pairs(nums2)
end = time.time()

print("\nReverse Pairs:", pairs)
print("Execution Time:", end-start, "seconds")


print("\nAlphabet Pattern")
start = time.time()
alphabet_pattern(4)
end = time.time()
print("Execution Time:", end-start, "seconds")
```
